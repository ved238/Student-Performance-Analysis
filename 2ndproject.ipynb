```python
mp.figure(figsize=(8,6))
grouped = a.groupby("Attendence Category")["Total_Score"].mean()

mp.bar(grouped.index, grouped)
mp.xlabel("Attendence")
mp.ylabel("marks")
mp.title("Attendence vs Marks")
 
mp.show()
```

```python
grouped2 = a.groupby("Grade")["Student_ID"].count()

#mp.pie(grouped2, labels=grouped2.index, autopct="%1.1F%%")
mp.figure(figsize=(12,6))

mp.subplot(1, 2, 1)
mp.bar(grouped2.index, grouped2)
mp.title("Grade VS No.of Students")
mp.xlabel("Grade")
mp.ylabel("Total Students")


mp.subplot(1, 2, 2)
mp.plot(grouped2.index, grouped2)
mp.title("Grade VS No.of Students")
mp.xlabel("Grade")
mp.ylabel("Total Students")

mp.show()
```
